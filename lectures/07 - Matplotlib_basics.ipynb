```python
import matplotlib.pyplot as plt
import numpy as np

N = 4

survived = (70, 80, 20, 300 )
perished = (20, 30, 319, 100)


ind = np.arange(N)    # the x locations for the groups
width = 0.35       # the width of the bars: can also be len(x) sequence

fig, ax = plt.subplots()

p1 = ax.bar(ind, survived, width, label='Survived')

p2 = ax.bar(ind, perished, width,
            bottom=survived, label='Perished')

ax.axhline(0, color='grey', linewidth=1)
ax.set_ylabel('Count')
ax.set_title('Survival by group and gender')

ax.set_xticks(ind, labels=['Boys', 'Girls', 'Men', 'Women'])
ax.legend()

# Label with label_type 'center' instead of the default 'edge'
ax.bar_label(p1, label_type='center')
ax.bar_label(p2, label_type='center')
ax.bar_label(p2)

plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

plt.style.use('_mpl-gallery')

# make data:
np.random.seed(3)

x = 0.5 + np.arange(8)
y = np.random.uniform(2, 7, len(x))

# plot
fig, ax = plt.subplots()

ax.bar(x, y, width=1, edgecolor="white", linewidth=0.7)

ax.set(xlim=(0, 8), xticks=np.arange(1, 8),
       ylim=(0, 8), yticks=np.arange(1, 8))



plt.show()
```

# Groupby 

Like SQL groupBy - aggregating data of rows in a column into a cell

![groupby](https://miro.medium.com/max/1400/0*m-puRNwcnUl5zZ4O)

```python
# Below is a data dictionary - where each key in the dict refers to an array of data

data = {'Company': ['GOOG', 'GOOG', 'MSFT', 'MSFT', 'FB', 'FB'],
        'Person': ['Sam', 'Charlie', 'Amy', 'Vanessa', 'Carl', 'Sarah'],
        'Sales': [200, 120, 340, 124, 243, 350]
       }
```

```python
df = pd.DataFrame(data)
df
```

```python
byComp = df.groupby('Company')

#call aggregate function
byComp.mean()
```

```python
byComp.sum()
```

```python
byComp.sum().loc['FB']
```

# Merge

Similar to joins in SQL

```python
left = pd.DataFrame({'key': ['K0','K1','K2','K3'],
                     'A'  : ['A0','A1','A2','A3'],
                     'B'  : ['B0','B1','B2','B3']})

right = pd.DataFrame({'key': ['K0','K1','K2','K3'],
                     'C'  : ['C0','C1','C2','C3'],
                     'D'  : ['C0','C1','C2','C3']})
```

```python
left
```

```python
right
```

# Inner join

the intersect 

![innerjoin](https://learn.microsoft.com/en-us/power-query/images/inner-join-operation.png)


```python
pd.merge(left, right, how= 'inner', on='key')
```

```python
left = pd.DataFrame({'key1': ['K0','K0','K1','K2'],
                     'key2': ['K0','K1','K0','K1'],
                      'A'  : ['A0','A1','A2','A3'],
                      'B'  : ['B0','B1','B2','B3']})

right = pd.DataFrame({'key1': ['K0','K1','K1','K2'],
                      'key2': ['K0','K0','K0','K0'],
                       'C'  : ['C0','C1','C2','C3'],
                       'D'  : ['C0','C1','C2','C3']})
```

```python
pd.merge(left, right, on=['key1','key2'])
```

# Outer join

![outerjoins](https://www.ionos.co.uk/digitalguide/fileadmin/DigitalGuide/Screenshots_2018/Outer-Join.jpg)

```python
pd.merge(left, right, how= 'outer', on=['key1','key2'])
```

```python
pd.merge(left, right, how= 'right', on=['key1','key2'])
```

```python
pd.merge(left, right, how= 'left', on=['key1','key2'])
```

# Joining 

Similar to merge, but will join on the index, rather than the column



```python
left = pd.DataFrame({'A'  : ['A0','A1','A2','A3'],
                     'B'  : ['B0','B1','B2','B3']},
                    index = ['K0','K1','K2','K3'])

right = pd.DataFrame({'C'  : ['C0','C1','C2','C3'],
                     'D'  : ['C0','C1','C2','C3']},
                    index = ['K0','K1','K2','K3'])
```

```python
left.join(right)
```
